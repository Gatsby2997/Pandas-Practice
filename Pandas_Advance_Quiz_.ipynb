```python
import pandas as pd
```

```python
data = pd.DataFrame({'A' : [1,2,3], 'B': [4,5,6]})
```

```python
data
```

```python
new_index = ['x', 'y', 'z']
df = data.reindex(new_index)
```

```python
df
```

```python
df1 = {'name':['John', 'Jane', 'Bob'], 'age' : [25, 30, 35]}
```

```python
df1 = pd.DataFrame(df1)
```

```python
df1
```

```python
df2 = df1.reindex([2,1,0])
```

```python
for index, row in df1.iterrows():
  print(index)
```

```python
df2['name_upper'] = df1['name'].str.upper()
```
