Write a code to find the number of passenger who is below the avg age ?

1.   Below avg age passenger
2.   Above avg age passenger
3.   Min Age
4.   Max Age


```python
count_below_avg_age = (df['Age'] < df['Age'].mean()).sum()
count_above_avg_age = (df['Age'] > df['Age'].mean()).sum()
min_age = df['Age'].min()
max_age = df['Age'].max()
```

```python
print("Number of passenger below avg age:", count_below_avg_age)
print("Number of passenger above avg age:", count_above_avg_age)
print("Minimum age of passenger:", min_age)
print("Maximum age of passenger:", max_age)
```
